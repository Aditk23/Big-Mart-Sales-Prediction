```python
train['train_test']=0
test['train_test']=1
test['Item_Outlet_Sales'] = np.NaN
all_data = pd.concat([train,test])
all_data
```

```python
all_data = all_data.reset_index(drop=True)
```

```python
all_data.loc[all_data.Item_Identifier=='FDA15'].Item_Weight
```

```python
missing = all_data.loc[pd.isnull(all_data.Item_Weight)].index
for i in missing:
    ii = all_data.iloc[i].Item_Identifier
    value = all_data.loc[all_data.Item_Identifier==ii].Item_Weight.mean()
    all_data.loc[all_data.index==i,'Item_Weight'] = value
```

```python
all_data.loc[all_data.Item_Identifier.str.startswith('FD'),'Item_Category'] = 'Food'
all_data.loc[all_data.Item_Identifier.str.startswith('DR'),'Item_Category'] = 'Drink'
all_data.loc[all_data.Item_Identifier.str.startswith('NC'),'Item_Category'] = 'Non Consumable'
```

```python
all_data.loc[all_data.Item_Category == 'Non Consumable','Item_Fat_Content'] = 'Not Edible'
```

```python
all_data.Item_Fat_Content = all_data.Item_Fat_Content.replace({'reg':'Regular','LF':'Low Fat','low fat':'Low Fat'})
```

```python
all_data['Age'] = 2020 - all_data['Outlet_Establishment_Year']
```

```python
all_data['MRP_per_weight'] = all_data['Item_MRP']/all_data['Item_Weight']
```

```python
all_data.loc[all_data.Item_MRP <=70 ,'MRP_cluster'] = 1
all_data.loc[(all_data.Item_MRP >70) & (all_data.Item_MRP <=140) ,'MRP_cluster'] = 2
all_data.loc[(all_data.Item_MRP >140) & (all_data.Item_MRP <=200) ,'MRP_cluster'] = 3
all_data.loc[all_data.Item_MRP >200 ,'MRP_cluster'] = 4
```
